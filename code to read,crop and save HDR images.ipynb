```python
import numpy as np
import imageio as io
import imageio
import matplotlib.pyplot as plt
```

### Read HDR image

```python
hdr=io.imread("pfstools_hdr_images/hdr_gallery/forest_path.hdr",format='HDR-FI')
hdr=np.array(hdr)
```

```python
plt.figure()
plt.imshow(hdr)
plt.show()
```

### Crop image

```python
hdr.shape
```

```python
w_start=10
w_end=500
h_start=20
h_end=550
```

```python
crop_hdr=hdr[w_start:w_end,h_start:h_end,:]
```

```python
crop_hdr.shape
```

```python
plt.figure()
plt.imshow(crop_hdr)
plt.show()
```

### Saving HDR image

```python
io.imwrite("crop_image.hdr",crop_hdr,format='HDR-FI')
```
